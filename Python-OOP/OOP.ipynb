```python
class Phone:
    category = 'Electronics'

    # Constructor method to initialize instance attributes
    def __init__(self, model, battery, camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # method to simulate using the phone and draining the battery
    def charge(self, hour):
        self.charge += hour

    def capture(self, photo):
        if(self.battery_percentage) <= 0:
            print("Battery is empty. Please charge the phone.")
        else:
            self.battery_percentage -= photo
            print(f"Captured a photo in {self.model}. Battery percentage is now {self.battery_percentage}%.")

# Create instances of the Phone class with specific attributes
apple = Phone('iPhone 12', '2815 mAh', '12 MP')
blackberry = Phone('BlackBerry Key2', '3500 mAh', '12 MP')
motorola = Phone('Moto G Power', '5000 mAh', '16 MP')

# Simulate using the phones and capturing photos
apple.capture(10)  # Output: Captured a photo in iPhone 12. Battery percentage is now 90%.
print(apple.battery_percentage)  # Output: 90
```

```python
# Inheritance in Python allows a new class (called a child class) to inherit attributes and methods from an existing class (called a parent class). This promotes code reusability and establishes a natural hierarchical relationship between classes.
# Single Inheritance
# Multiple Inheritance

class Phone:
    category = 'Electronics'

    # Constructor method to initialize instance attributes
    def __init__(self, model, battery, camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # method to simulate using the phone and draining the battery
    def charge(self, hour):
        print(f"Charge completed by {hour} hours on {self.model}")

    def capture(self, photo):
        if(self.battery_percentage) <= 0:
            print("Battery is empty. Please charge the phone.")
        else:
            self.battery_percentage -= photo
            print(f"Captured a photo in {self.model}. Battery percentage is now {self.battery_percentage}%.")


# Single Inheritance: A child class inherits from a single parent class.
class Smartphone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor

    def charge(self, hour):
        print(f"Smartphone {self.model} is charging for {hour} hours.")
        super().charge(hour)

pixel = Smartphone('Google Pixel 5', '4080 mAh', '12.2 MP', 'Snapdragon 765G')
print(pixel.model)  # Output: Google Pixel 5
print(pixel.processor)  # Output: Snapdragon 765G
pixel.charge(2)  # Simulate charging the phone for 2 hours
class Cooling_Mechanism:
    def __init__(self, cooling_method):
        self.cooling_method = cooling_method
    
    # method to simulate cooling the phone
    def cooling_on(self):
        print(f"Cooling mechanism {self.cooling_method} is now active.")

# Multiple Inheritance: A child class inherits from more than one parent class.
class SmartPhone_CoolingMode(Smartphone, Cooling_Mechanism):
    def __init__(self, model, battery, camera, processor, cooling_method):
        Smartphone.__init__(self, model, battery, camera, processor)
        Cooling_Mechanism.__init__(self, cooling_method)

    def charge(self, hour):
        print(f"Smartphone {self.model} is charging for {hour} hours with cooling mode {self.cooling_method}.")
        super().charge(hour)

galaxy = SmartPhone_CoolingMode('Samsung Galaxy S21', '4000 mAh', '64 MP', 'Exynos 2100', 'Liquid Cooling')
print(galaxy.model)  # From Smartphone class: Samsung Galaxy S21
print(galaxy.cooling_method)  # From Cooling_Mechanism class: Liquid Cooling
print(galaxy.processor)  # From Smartphone class: Exynos 2100
print(galaxy.battery)  # From Phone class: 4000 mAh
print(galaxy.cooling_on())  # From Cooling_Mechanism class: Cooling mechanism Liquid Cooling is now active.
print(galaxy.charge(3))  # From SmartPhone_CoolingMode class: Smartphone Samsung Galaxy S21 is charging for 3 hours with cooling mode Liquid Cooling. Charge completed by 3 hours on Samsung Galaxy S21
```

```python
# Polymorphism in Python allows objects of different classes to be treated as objects of a common superclass. It is the ability of different classes to respond to the same method call in different ways. This promotes flexibility and extensibility in code design.

class Camera:
    def __init__(self, name):
        self.name = name
    
    # method
    def capture(self):
        print(f"{self.name} captured a photo.")

class DSLR(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"{self.name} captured a photo with DSLR quality.")

class Smartphone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"{self.name} captured a photo with smartphone quality.")

class Drone(Camera):
    def __init__(self, name, resolution):
        super().__init__(name)
        self.resolution = resolution

    # method overriding
    def capture(self):
        print(f"{self.name} captured a photo with drone quality.")

# Create instances of each class
dslr_camera = DSLR('Canon EOS 5D', '30.4 MP')
smartphone_camera = Smartphone('iPhone 12', '12 MP')
drone_camera = Drone('DJI Mavic Air 2', '48 MP')

dslr_camera.capture()  # Output: Canon EOS 5D captured a photo.
smartphone_camera.capture()  # Output: iPhone 12 captured a photo.
drone_camera.capture()  # Output: DJI Mavic Air 2 captured a photo.
```

```python
# Encapsulation in Python is the concept of bundling data (attributes) and methods (functions) that operate on the data into a single unit, typically a class. It also involves restricting direct access to some of an object's components, which is a means of preventing accidental interference and misuse of the data. This is often achieved using private attributes and methods.

class Phone:
    def __init__(self, name, model, imei):
        self.name = name
        self.__model = model  # Private attribute
        self.__imei = imei  # Private attribute

    # Public method to access the private attribute
    def get_imei(self):
        return self.__imei

    # Public method to access the private attribute
    def get_model(self):
        return self.__model
    # Public method to set the private attribute
    def set_model(self, model):
        self.__model = model
    
phone = Phone('iPhone 12', 'A2172', '123456789012345')
print(phone.name)  # Output: iPhone 12
print(phone.get_model())  # Output: A2172
phone.set_model('A2173')
print(phone.get_model())  # Output: A2173

# print(phone.__imei)  # This will raise an AttributeError because __imei is a private attribute
print(phone.get_imei())  # Output: 123456789012345
```

```python
# Abstraction in Python is the concept of hiding the complex implementation details of a system and exposing only the necessary and relevant parts to the user. It allows developers to focus on what an object does rather than how it does it. This is often achieved using abstract classes and methods.
from abc import ABC, abstractmethod # Importing the ABC module to create abstract classes

class Telephone(ABC):
    @abstractmethod
    def make_call(self):
        pass

class Smartphone(Telephone):
    def make_call(self):
        print("Making a call using the smartphone.")

class IPhone(Telephone):
    def make_call(self):
        print("Making a call using the iPhone.")

smartphone = Smartphone()
iphone = IPhone()
smartphone.make_call()  # Output: Making a call using the smartphone.
iphone.make_call()  # Output: Making a call using the iPhone.
```
