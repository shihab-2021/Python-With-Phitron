```python
class Phone:
    category = 'Electronics'

# Create instance of the Phone class
apple = Phone()
blackberry = Phone()
motorola = Phone()

print(apple.category)  # Output: Electronics
print(blackberry.category)  # Output: Electronics
print(motorola.category)  # Output: Electronics

# Change the category for the Phone class
apple.category = 'Mobile Devices'
print(apple.category)  # Output: Mobile Devices
```
